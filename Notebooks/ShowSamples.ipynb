```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
images_file = r"j:\Temp\resized_images.npy"
predictions_file = r"j:\Temp\predictions.npy"

classes = ['A', 'C', 'None', 'R', 'S']
```

```python
resized_images = np.load(images_file)
predictions = np.load(predictions_file)
```

```python
num_choices = 5
n = resized_images.shape[0]
n_indices = range(n)

random_choice = np.random.choice(n_indices, num_choices)

plt.figure(figsize=(3*num_choices, 3))

for i in range(num_choices):
    im = resized_images[random_choice[i]][:,:,:]
    prediction = predictions[random_choice[i]]
    plt.subplot(1, num_choices, i+1)
    max_index = np.argmax(prediction)
    plt.title(str(classes[max_index]) + ": " + str(round(prediction[max_index],2)) + "%")
    plt.imshow((im + 1)/2)
    plt.axis('off')
```
